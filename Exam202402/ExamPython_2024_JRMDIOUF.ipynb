```python
class Curve:
    def __init__(self, start, stop, nbr_points=5432) :
        self.__start = start
        self.__stop = stop
        self.__nbr_points = nbr_points
    
    def __get_image(self,x):
        return x**5
    
    def trace_curve(self):
        X = np.linspace(self.__start,self.__stop,50)
        Y = self.__get_image(X)
        plt.figure(figsize=(10,10))
        plt.grid(visible=True)
        plt.plot(X, Y, color='r')
        #plt.show()

    def get_random_floats(self, debut, fin):
        floats = []
        for i in range(self.__nbr_points):
            rand_float = random.uniform(debut, fin)
            while(rand_float in floats) :
                 rand_float = random.uniform(debut, fin)
            floats.append(rand_float)
        return floats

    def is_point_over_curve(self, x):
        return self.__get_image(x[0]) < x[1]

    def get_green_surface(self):
        X = np.linspace(self.__start,self.__stop,100)
        Y = self.__get_image(X)
        area = 0
        for i in range(1, len(X)-1):
            
            area+=np.trapz(X,Y,1)
        return area
    def display_random_points(self):
        X = self.get_random_floats(self.__start, self.__stop)
        Y = self.get_random_floats(self.__get_image(self.__start)
                                   , self.__get_image(self.__stop))

        for x,y in zip(X,Y):
            clr = 'g'
            mrk = '.'

            if self.is_point_over_curve([x,y]):
                clr='b'
                mrk='x'

            plt.scatter(x,y, marker=mrk, c=clr)
    
    @property
    def start(self):
        return self.__start

    @start.setter
    def start(self, val):
        if not isinstance(val, float):
            raise ValueError("Valeur non permise pour cette propriété. Nombre décimal attendu")
        self.__start = val
    
    @property
    def stop(self):
        return self.__stop

    @stop.setter
    def stop(self, val):
        if not isinstance(val, float):
            raise ValueError("Valeur non permise pour cette propriété. Nombre décimal attendu")
        self.__stop = val

    @property
    def nbr_points(self):
        return self.__nbr_points

    @nbr_points.setter
    def nbr_points(self, val):
        if not isinstance(val, float):
            raise ValueError("Valeur non permise pour cette propriété. Nombre décimal attendu")
        self.__nbr_points = val
```

```python
crv = Curve(0,1,100)

crv.trace_curve()
crv.display_random_points()

plt.show()

crv.get_green_surface()
```
